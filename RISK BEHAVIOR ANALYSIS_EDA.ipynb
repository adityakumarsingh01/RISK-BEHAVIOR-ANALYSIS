OUTLIERS
===

```python
# Loop for checking outliers
for col in data.select_dtypes(include=['int64', 'float64']):
    Q1 = data[col].quantile(0.25)
    Q3 = data[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    
    outliers = data[(data[col] < lower) | (data[col] > upper)]
    
    if not outliers.empty:
        print(f"{col}: {len(outliers)} outliers")
```

```python
# Box plot for Outliers 
# Columns to check for outliers
outliers_cols = ['smokes_per_day', 'drinks_per_week', 'sleep_hours', 'bmi']

# Create plots
plt.figure(figsize = (16, 5))
colors = ['skyblue', 'lightgreen', 'salmon', 'gold']

for i, col in enumerate(outliers_cols):
    plt.subplot(1, 4, i+1)
    sns.boxplot(y = data[col], color = colors[i])
    plt.title(f'Outliers in {col}')

plt.show()
```

```python
# Replace Outliers with Mean
def rep_out(data, column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    mean = data[column].mean()
    data[column] = data[column].apply(lambda x: mean if x < lower or x > upper else x)
    return data

data = rep_out(data, outliers_cols[0])
data = rep_out(data, outliers_cols[1])
data = rep_out(data, outliers_cols[2])
data = rep_out(data, outliers_cols[3])
```

```python
colors = ['mediumorchid', 'deepskyblue', 'tomato', 'khaki']

plt.figure(figsize = (16, 5))

for i, col in enumerate(outliers_cols):
    plt.subplot(1, 4, i+1)
    sns.boxplot(y = data[col], color = colors[i])
    plt.title(f'Outliers in {col}')

plt.show()
```
